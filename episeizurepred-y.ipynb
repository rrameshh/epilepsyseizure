```python
data_path = "/kaggle/input/epileptic-seizure-recognition/Epileptic Seizure Recognition.csv"
import pandas as pd
import os
df=pd.read_csv(data_path)
df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Flatten
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
y = df['y']
df = df.drop(['Unnamed'], axis=1)

```

```python
df = df.drop(['y'], axis=1)
X = []
for i, row in df.iterrows():
    tmp_x = []
    for elem in row.tolist():
         tmp_x.append([elem])
    X.append(tmp_x)
```

```python
for i in range(y.shape[0]):
    if y[i]!=1:
        y[i]=0
```

```python
X = np.array(X)
```

```python
# Scaling the data
#scaler = StandardScaler()
#X_scaled = scaler.fit_transform(X)

model = Sequential()
model.add(LSTM(64, input_shape=(178,1), return_sequences=True))
model.add(LSTM(32))
model.add(Flatten())
model.add(Dense(250, activation='relu'))

model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model.fit(X_train, y_train, validation_data=(X_test,y_test), epochs=100, batch_size=128, verbose=2)
```
